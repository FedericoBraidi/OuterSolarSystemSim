```python
import numpy as np
import matplotlib.pyplot as plt
from time import process_time,time
from KeplerOuter import KeplerOuter
```

```python
#METODO DI EULERO ALL'AVANTI

if __name__ == "__main__":
    # parametri del run
    hEA = [10,1,0.1]
    neuler = [20,200,2000]
    NEA = 1250
    for j in range(len(hEA)):
        EA = KeplerOuter(N=NEA,h=hEA[j])
        tic = process_time()
        for i in range(NEA):
            EA.forward_euler(i,neuler=neuler[j])
        print(" cpu time for cycle t=", process_time()-tic)
        fig = plt.figure(figsize = (9,7))
        ax = fig.add_subplot(111, projection = '3d')
        ax.set_xlabel("x")
        ax.set_ylabel("y")
        ax.set_zlabel("z")
        ax.plot(EA.qx[:,1]-EA.qx[:,0], EA.qy[:,1]-EA.qy[:,0], EA.qz[:,1]-EA.qz[:,0])
        ax.plot(EA.qx[:,2]-EA.qx[:,0], EA.qy[:,2]-EA.qy[:,0], EA.qz[:,2]-EA.qz[:,0])
        ax.plot(EA.qx[:,3]-EA.qx[:,0], EA.qy[:,3]-EA.qy[:,0], EA.qz[:,3]-EA.qz[:,0])
        ax.plot(EA.qx[:,4]-EA.qx[:,0], EA.qy[:,4]-EA.qy[:,0], EA.qz[:,4]-EA.qz[:,0])
        ax.plot(EA.qx[:,5]-EA.qx[:,0], EA.qy[:,5]-EA.qy[:,0], EA.qz[:,5]-EA.qz[:,0])
        ax.plot(EA.qx[:,0],EA.qy[:,0], EA.qz[:,0])
        fig.savefig("fe_neu_%u.png" %neuler[j])
        plt.show()
```

```python
#METODO RK4

NRK4 = 3500
nrun_kut_4 = [2,20,200,2000]
tic = process_time()
hrk4 = [100,10,1,0.1]
for j in range(len(hrk4)):
    RK4 = KeplerOuter(N=NRK4,h=hrk4[j])
    for i in range(NRK4):
        RK4.RK4(i,nrk4=nrun_kut_4[j])
    print(" cpu time for cycle t=", process_time()-tic)
    fig = plt.figure(figsize = (9,7))
    ax = fig.add_subplot(111, projection = '3d')
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    ax.plot(RK4.qx[:,1]-RK4.qx[:,0], RK4.qy[:,1]-RK4.qy[:,0], RK4.qz[:,1]-RK4.qz[:,0],label="Giove")
    ax.plot(RK4.qx[:,2]-RK4.qx[:,0], RK4.qy[:,2]-RK4.qy[:,0], RK4.qz[:,2]-RK4.qz[:,0],label="Saturno")
    ax.plot(RK4.qx[:,3]-RK4.qx[:,0], RK4.qy[:,3]-RK4.qy[:,0], RK4.qz[:,3]-RK4.qz[:,0],label="Urano")
    ax.plot(RK4.qx[:,4]-RK4.qx[:,0], RK4.qy[:,4]-RK4.qy[:,0], RK4.qz[:,4]-RK4.qz[:,0],label="Nettuno")
    ax.plot(RK4.qx[:,5]-RK4.qx[:,0], RK4.qy[:,5]-RK4.qy[:,0], RK4.qz[:,5]-RK4.qz[:,0],label="Plutone")
    ax.plot(RK4.qx[:,0],RK4.qy[:,0], RK4.qz[:,0],label="Sole")
    fig.savefig("rk4_nrk_%u" %nrun_kut_4[j])
    plt.show()
```
